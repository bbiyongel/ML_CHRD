## 3. 최소 최대 탐색

# 데이터의 부분집합 선택

## 1. 데이터 선택

### 1.1. 인덱서(Indexer)
 - 인덱싱 : 데이터프레임에서 특정한 데이터만 골라내는 것
 - 2차원(numpy 행렬) 인덱싱을 지원하기 위해 인덱서(indexer) 속성 제공
 - loc : 라벨값 기반의 2차원 인덱싱
 - iloc : 순서를 나타내는 정수 기반의 2차원 인덱싱

### loc 인덱서

```python
# df.loc[행 인덱싱값]
# df.loc[행 인덱싱값, 열 인덱싱값]
df = pd.DataFrame(np.arange(10, 22).reshape(3,4),
                 index=["a","b","c"],
                 columns=["A","B","C","D"])
df
```

```python
# 인덱싱값을 하나만 받는 경우
df.loc["a"]
# "a"인 행을 고르면 해당하는 행이 시리즈로 출력
```

```python
df.loc["b":"c"]
```

```python
df["b":"c"]
```

```python
# 인덱스데이터의 리스트
df.loc[["b", "c"]]
# 이 때는 loc를 쓰지 않으면 KeyError 오류가 발생
```

```python
df.A > 15
```

```python
df.loc[df.A > 15]
```

```python
df.loc["A"] # KeyError
```

```python
df.loc["A", "B"] # KeyError
```

```python
# 인덱싱값으로 라벨 데이터의 슬라이싱 또는 리스트 사용 가능
df.loc["b":, "A"]
```

```python
df.loc[["a","b"], ["B", "D"]]
```

### iloc 인덱서
- iloc 인덱서는 loc 인덱서와 반대로 라벨이 아니라 순서를 나타내는 
- 정수(integer) 인덱스만 받는다. 다른 사항은 loc 인덱서와 같다.

```python
df.iloc[0,1]
```

```python
df.iloc[:2, 2]
```

```python
df.iloc[0, -2:]
```

```python
df.iloc[2:3, 1:3]
```

```python
## 변수형
### 범주형 변수 (Categorical & Qualitative)
### 1) 명목형 (Nominal : 명목변수)
- 객체간 서로 구분하기 위한 이름들.
- 순서가 의미가 없다.
- 예시) zip codes, 성별, 사원번호, 눈색깔 등

### 2) 순위형 (Ordinal : 서열변수)
- 순서의 의미가 있는 범주를 갖는 자료.
- 예시) 광물강도, 성적, 제품 선호도 등
```

```python
### 수치형 변수 (Numeric & Quantitative)
### 1) 정수형 (Discrete & Interval : 등간형)
- 셀 수 있으며, 간격이 일정함.
- 0의 의미는 없다.
- 예시) 달력일자, 온도(섭씨, 화씨) 등
### 2) 연속형 (Continuous & Ratio : 비율형)
- 연속적인 값을 가진다.
- 차이(Difference)와 비율(Ratio)이 의미있음.
- 예시) 온도(절대온도), 판매량, 비용, 연령, 길이 등
```
