```python
def train_plot(col_num):
    train[train.gender==0].iloc[:, col_num].hist(figsize=(10,10), color='red', label = 'Female')
    train[train.gender==1].iloc[:, col_num].hist(figsize=(10,10), label = 'Male')
    plt.title(train.columns[col_num])
    plt.legend()
    plt.show()
```

```python
for num in range(2, len(train.T)):
    train_plot(num)
```

```python
def plot(X): # 차트 만들기 함수
    male = train[train['gender']==1][X].value_counts()
    # train의 'gender'열의 값이 1(남성)인 것들을 세서 male 구성
    female = train[train['gender']==0][X].value_counts() 
    # train의 'gender'열의 값이 0(여성)인 것들을 세서 female 구성
    df = pd.DataFrame([male,female]) # male, female data로 이루어진 df 구성
    df.index = ['male','female'] # df의 인덱스 설정
    df.plot(kind='bar',stacked=True, figsize=(10,5)) # plot 종류, 위로 쌓은 것, plot 사이즈 고정
```

```python
plot('주구매코너')
```

## 그냥 예측 해보자

```python
c_train = train.copy()
```

```python
c_train.dropna(axis=0, inplace=True)
```

```python
len(c_train)
```

```python
c_train.isnull().sum()
```

```python
c_train = pd.get_dummies(c_train, columns=['주구매코너'])
c_train.filter(like='주구매코너').head()
```

```python
from sklearn.model_selection import train_test_split # for Hold-out validation
dfX = c_train.drop(['custid', 'gender'], axis=1) # exclude 'custid' attribute & class variable
dfy = c_train['gender']
X_train, X_test, y_train, y_test = train_test_split(dfX, dfy, \
                                                    test_size=0.25, random_state=0)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=6, random_state=0)
tree.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
r_forest = RandomForestClassifier()
r_forest.fit(X_train, y_train)
```

```python
display(tree.score(X_train, y_train))
display(tree.score(X_test, y_test))
```

```python
display(r_forest.score(X_train, y_train))
display(r_forest.score(X_test, y_test))
```
